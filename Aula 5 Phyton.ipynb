1. Escreva uma rotina que crie uma borda ao redor de uma palavra para destacá-la como sendo um título. A rotina deve receber como Parâmetro a palavra a ser destavada. o tamanho da caixa de texto deveráser adptável de acordo com o tamanho da palavra.

```python
#Definir função titulo:
def titulo (nome):
  print('*','-' * len(nome),'*')
  print('|',nome,'|')
  print('*','-' * len(nome),'*')

nome = input('Incira o nome do titulo: ')
if len(nome):
  titulo(nome)
```

```python
#Exercício 1
def borda (s1):
  tam = len(s1)
  # só imprime caso exista algum caractere:
  if tam:  # se tam for verdadeiro, tem algum caractere.
    print('+','-' * tam,'+')
    print('|',s1,'|')
    print('+','-' * tam,'+')

#Programa principal
borda('Olá mundo')
borda(input('escreva algum nome para borda'))
```

#|Exercício


Escreva uma função para validar uma string. Essa função recebe como parâmetro a string, o número mínimo e máximo de caracteres.
retorne verdadeiro se o tamanho da string estiver entre os valores mínimo e máximo, e falso, caso contrario.

```python
#Teste com repetição englobando solicitado do exercicio
def validar (s1):
  tam = len(s1)
  if tam <=3 or tam >=9:
    valid = False
  else:
    valid = True
  return valid

while True:
  s1 = input('Escreva um nome com 4 a 8 caracteres: ')
  valid = validar(s1)
  if valid == True:
    print('Nome valido!')
    break
  else:
    print('nome invalido, tente novamente!')
    continue
```

```python
#Exercicio 1
def validar_string (pergunta, min, max):
  s1 = input(pergunta)
  tam = len(s1)
  while ((tam < min) or (tam > max)):
    s1 = input(pergunta)
    tam = len(s1)
  return s1

#Programa principal
x = validar_string('Digite uma string: ',10,30)
print('você digitou a string: {}. \n Dado válido. Encerrando o programa...'.format(x)) # o \n pode ser usado para escrever o restante do print na linha de baixo
```

```python
while True:
  try:
    x = int(input('Por favor digite um número: '))
    break
  except ValueError:
    print('oops! Número invalido. \n Tente novamente...')
```

```python
def div():
  try:
    num1 = int(input('Digite um número: '))
    num2 = int(input('Digite um número: '))
    res = num1 / num2
  except ZeroDivisionError:  # se acontecer erro de zero  <-
    print('OPS! Erro de divisão por zero...')
  except:   # se acontecer qualquer outro erro <-
    print('Algo de errado não está certo...')
  else: # se não der erro
    return res
  finally: #sempre acontece mesmo com erro ou não
    print('sempre vai executa!')

#Programa principal
res = div()
print(res)
```

# Função com parâmetro de função

```python
def imprime_com_condicao (num,fcond):
  if fcond(num):
    print(num)

def par(x):
  return x % 2 == 0
def impar(x):
  return not par(x)

#Programa inicial.
imprime_com_condicao(2,par)
```
