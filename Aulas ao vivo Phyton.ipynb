# Composição

```python
nota = 8.5
s1 = 'Você tirou %f na disciplina de Algoritimos' % nota
print (s1)
```

Limitando casas decimais


```python
nota = 8.5
s1 = 'Você tirou %.2f na disciplina de Algoritimos' % nota
print (s1)
```

Várias variáveis


```python
nota = 8.5
disciplina = 'Algoritmos'
s1 = 'Você tirou %f na disciplina de %s' % (nota, disciplina)
print (s1)
```

```python
s1 = 'Você tirou {} na disciplina {}' .format(nota,disciplina)
print(s1)
```
