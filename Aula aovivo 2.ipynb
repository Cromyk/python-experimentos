```python
#exercicio 2 de 4 da atividade pratica
print('Bem-vindo ao sistema de pizza do Marco Antônio de Souza Matos (EU: 4316762)')
print(' ----------------------------- CARDÁPIO ------------------------------')
print('|   Código  |   Descrição   |   Pizza média(M)  |   Pizza Grande(G)   |')
print('|     21    |   Napolitano  |      R$20,00      |        R$26,00      |')
print('|     22    |   Margherita  |      R$20,00      |        R$26,00      |')
print('|     23    |   Calabresa   |      R$25,00      |        R$32,50      |')
print('|     24    |   Toscana     |      R$30,00      |        R$39,00      |')
print('|     25    |   Portuguesa  |      R$30,00      |        R$39,00      |')
print(' ---------------------------------------------------------------------')

c21m = 0
c22m = 0
c23m = 0
c24m = 0
c25m = 0
c21g = 0
c22g = 0
c23g = 0
c24g = 0
c25g = 0
soma=0
while True:
  tam = input('Insira o tamanho da pizza entre M e G: ')
  if tam != 'M'and tam != 'm' and tam != 'g' and tam != 'G':
    print('Tamanho invalido, tente novamente')
    continue
  cod = int(input('Digite o codigodo produto: '))
  if (cod != 21) and (cod != 22) and (cod != 23) and (cod != 24) and (cod != 25):
    print('produto invalido, tente novamente.')
    continue
#pizza 21
  if cod == 21:
    if tam == 'M' or tam == 'm':
      soma += 20
      c21m += 1
    elif tam == 'G' or tam == 'g':
      soma += 26
      c21g += 1
  #pizza 22
  elif cod == 22:
    if tam == 'M' or tam == 'm':
       soma += 20
       c22m += 1
    elif tam == 'G' or tam == 'g':
       soma += 26
       c22g += 1
#pizza 23
  elif cod == 23:
      if tam == 'M' or tam == 'm':
        soma += 25
        c23m += 1
      elif tam == 'G' or tam == 'g':
        soma += 32.5
        c23g += 1
#pizza 24
  elif cod == 24:
      if tam == 'M' or tam == 'm':
        soma += 30
        c24m +=1
      elif tam == 'G' or tam == 'g':
        soma += 39
        c24g +=1
#pizza 25
  elif cod == 25:
      if tam == 'M' or tam == 'm':
        soma += 30
        c25m +=1
      elif tam == 'G' or tam == 'g':
        soma += 39
        c25g +=1
 
#    
  rep = input('Deseja pedir mais alguma pizza? Digite "S" para sim:')
  rep= rep.upper()  # com isso resolve problemas com maiusculo ou minusculo
  if rep == 's' or rep == 'S':
    continue
# pedido e total  
  else: 
    if c21m > 0:
      print('Você pediu {} pizzas Napolitana Média'.format(c21m))
    if c22m > 0:
      print('Você pediu {} pizzas Margherita Média'.format(c22m))
    if c23m > 0:
      print('Você pediu {} pizzas Calabresa Média'.format(c23m))
    if c24m > 0:
      print('Você pediu {} pizzas Toscana Média'.format(c24m))
    if c25m > 0:
      print('Você pediu {} pizzas Portuguesa Média'.format(c25m))
    if c21g > 0:
      print('Você pediu {} pizzas Napolitana Grande'.format(c21g))
    if c22g > 0:
      print('Você pediu {} pizzas Margherita Grande'.format(c22g))
    if c23g > 0:
      print('Você pediu {} pizzas Calabresa Grande'.format(c23g))
    if c24g > 0:
      print('Você pediu {} pizzas Toscana Grande'.format(c24g))
    if c25g > 0:
      print('Você pediu {} pizzas Portuguesa Grande'.format(c25g))

    print('Total do seu pedido é de : R${}'.format(soma))
    break
  
```
